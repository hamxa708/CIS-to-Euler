```python
import ipywidgets as widgets
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, latex
from IPython.display import Math,clear_output
```

```python
def cis_to_euler(real, imag):
    clear_output()
    magnitude = np.sqrt(real**2 + imag**2)

# Create the figure and axes
    fig = plt.figure(figsize=(3, 3))
    plt.xlim(-6-real, 6+real)  # Set x-axis limits from 0 to 6
    plt.ylim(-6-imag, 6+imag) 
# Plot the real and imaginary components
    plt.plot([0, real], [0, imag], 'ro-')
    plt.plot([-6-real,6+real],[0,0],'k--')
    plt.plot([0,0],[-6-imag,6+imag],'k--')

# Add labels and gridlines
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.title(f'Argand Diagram (Magnitude: {magnitude})')

    plt.show()
    r = (abs(imag) ** 2 + abs(real) ** 2) ** 0.5
    theta = np.arctan2(imag, real)
    if theta < 0:
        theta_title = 2 * np.pi + theta
    else:
        theta_title = theta

    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(121, polar=True)
    ax.plot([0, theta], [0, r])
    ax.plot(theta, r, 'ko')
    ax.plot(0, 0, 'ro')
    ax.set_title(r'$re^{i\theta}$ where $r=%.2f$ and $\theta=%.2f\pi$ or %.2f°' % (r, (theta_title / np.pi), np.degrees(theta_title)))
    plt.show()

real_input = widgets.FloatText(description='Real:')
imag_input = widgets.FloatText(description='Imaginary:')
calculate_button = widgets.Button(description='Calculate')

def calculate_euler(_):
    real = real_input.value
    imag = imag_input.value
    cis_to_euler(real, imag)

calculate_button.on_click(calculate_euler)

display(real_input, imag_input, calculate_button)
```
